```python
'''
获取新浪新闻主页下的全部新闻页
http://news.sina.cn

1. 获取首页下的全部新闻链接
2. 利用正则表达式解析界面得到url
3. 将链接保存到url集合中去
4. 依次从url set中获取url派去界面
5. 保存新闻信息到文本中
'''

import urllib3

class Downloader(object):
    "下载器类"
    
    def __init__(self):
        pass
    
    @staticmethod
    def download(url):
        '''
        下载器的静态方法
        @param url 链接的地址
        @return string 类型数据
        '''
        manager = urllib3.PoolManager()
        res = manager.request('GET', url)
        if res.status == 200:
            content = res.data.decode("utf-8")
            return content
        else:
            return res
        pass

    pass
```

```python
'''
界面分析器
用于将从下载器中得到的界面html代码转换成url集合
'''

import re
import json

class MainParser(object):
    
    def __init__(self):
        pass
    
    @staticmethod
    def parse(content):
        '''
        解析函数
        @content 从下载器中得到的内容
        @return url set url集合
        '''
        cotent = re.sub(r'[.\s\S]*？<section class="s_card j_tianyi" id="j_tianyi" data-sudaclick="tianyi_data">', '', content)
        print(content)        
        pass
    
    
    @staticmethod
    def parse2(content):
        '''
        将获取得到的字符串转换成json对象
        再转换成url集合
        @param content json字符串
        @return 
        '''
        obj = json.loads(content)
        data = obj['data']
        urlSet = set()
        for news in data:
            urlSet.add(news['url'])
            pass
#         print(urlSet)
        return urlSet
        pass
    pass

```

```python
url = 'http://cre.dp.sina.cn/api/v3/get?cateid=1o&cre=tianyi&mod=wnews&merge=3&statics=1&length=20&tm=1489716199' # tianyiApi
# url = 'https://news.sina.cn' # 直接抓取只能得到轮播图部分
res = Downloader.download(url=url)
print("download news lists success!")
if isinstance(res, urllib3.HTTPResponse): # 出错直接返回了结果
    print("Download failed: ", res.info())
    exit(0)
    pass
else:
#     print(res)
    res = res.replace(r'\/', r'/') # 去掉了文本中的转义字符
    f = open('sina_news.json', 'w', encoding='utf-8') # 写入到文件中保存
    f.write(res)
    f.flush()
    f.close()
    print("write news info json to file success!")
    urlSet = MainParser.parse2(res)
    for index, link in enumerate(urlSet):
        print(str(index), link)
        content = Downloader.download(link)
        if not isinstance(content, urllib3.HTTPResponse):
            # 写入文件
            f = open(str(index)+".html", 'w', encoding='utf-8')
            f.write(content)
            f.flush()
            f.close()
            print("write {0} file ok!".format(index))
        pass
    print("download all news success!")
    pass
```
